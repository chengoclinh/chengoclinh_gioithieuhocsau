btvn 01
cho y = 5x^5 + 6x^3 - 3x + 1
cho biet do doc cua da thuc tren o diem nao

```python
# BTVN 01
# Tính độ dốc (đạo hàm) của đa thức

import sympy as sp

# Khai báo biến
x = sp.symbols('x')

# Khai báo hàm số
y = 5*x**5 + 6*x**3 - 3*x + 1

# Tính đạo hàm (độ dốc)
dy_dx = sp.diff(y, x)

# In kết quả
print("Hàm số y =", y)
print("Độ dốc (đạo hàm) của hàm là:")
print("y' =", dy_dx)

# Tính độ dốc tại một số điểm minh họa
print("\nĐộ dốc tại x = 0:", dy_dx.subs(x, 0))
print("Độ dốc tại x = 1:", dy_dx.subs(x, 1))
print("Độ dốc tại x = -1:", dy_dx.subs(x, -1))
```

02
tao 1 tensor ban dau co gia tri 2
dinh nghia ham so va tinh gradient
y = x^3 + 2x^2 + 5x + 1
hay tinh dy/dx tai gia tri x
dung pp gradient descent voi lerning rate alpha = 0.1 de cap nhat giatri x trong 10 vong lap

```python
# BTVN 02

import torch

# Khởi tạo tensor x ban đầu
x = torch.tensor(2.0, requires_grad=True)

# Learning rate
alpha = 0.1

print("Giá trị x ban đầu:", x.item())

# Gradient Descent trong 10 vòng lặp
for i in range(10):
    # Định nghĩa hàm số
    y = x**3 + 2*x**2 + 5*x + 1

    # Tính gradient dy/dx
    y.backward()

    # Cập nhật x theo Gradient Descent
    with torch.no_grad():
        x -= alpha * x.grad

    # Reset gradient về 0
    x.grad.zero_()

    print(f"Vòng {i+1}: x = {x.item():.6f}, y = {y.item():.6f}")
```

03
tao 1 tap du lieu gia lap voi x la so gio (ngau nhien tu 1 - 10) va y la so diem dc tinh theo cong thuc y = 3x + 5 + noise
voi noise la gia tri ngau nhien nho

1. khoi tao tham so w va b ngau nhien voi require_grad = true
2. tinh mse
3. tinh gradient
4. cap nhat tham so w va b bang gradient  descent voi lerning rate alpha = 0.01
5. kap lai qua trinh tren trong vong lap de xem su hoi tu cua mo hinh

```python
# BTVN 03 - Giới thiệu về Học sâu

import torch

# 1. Tạo tập dữ liệu giả lập
torch.manual_seed(42)

x = torch.randint(1, 11, (20,), dtype=torch.float32)   # số giờ học
noise = torch.randn(20) * 0.5                          # nhiễu nhỏ
y = 3 * x + 5 + noise                                  # điểm số

# 2. Khởi tạo tham số w, b
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

# Learning rate
alpha = 0.01

# 3. Huấn luyện mô hình
epochs = 100

for epoch in range(epochs):
    # Dự đoán
    y_pred = w * x + b

    # 4. Tính MSE loss
    loss = torch.mean((y_pred - y) ** 2)

    # 5. Tính gradient
    loss.backward()

    # 6. Cập nhật tham số bằng Gradient Descent
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad

    # Reset gradient
    w.grad.zero_()
    b.grad.zero_()

    # In kết quả
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")
```

- Ở trường hợp thứ 2:
+ torch.tensor(arr) coppy dữ liệu sang vùng nhớ mới
+ Tensor x độc lập với numpy arr

```python
# BTVN 05 - Giới thiệu về Học sâu
# Tạo tensor: empty, zeros, ones, random, reshape

import torch

# 1. Tạo tensor empty
a = torch.empty(2, 3)
print("Tensor empty:")
print(a)

# 2. Tạo tensor zeros
b = torch.zeros(2, 3)
print("\nTensor zeros:")
print(b)

# 3. Tạo tensor ones
c = torch.ones(2, 3)
print("\nTensor ones:")
print(c)

# 4. Tạo tensor random
d = torch.rand(2, 3)
print("\nTensor random:")
print(d)

# 5. Reshape bằng view
e = torch.arange(0, 6)
print("\nTensor ban đầu:")
print(e)

e_view = e.view(2, 3)
print("\nReshape bằng view (2x3):")
print(e_view)

# 6. Reshape bằng as (reshape)
e_as = e.reshape(3, 2)
print("\nReshape bằng reshape (3x2):")
print(e_as)
```
