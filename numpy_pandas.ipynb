              [99,99,99],

hãy tưởng tượng ô 99 trống
giả xử 1 là x

lấy đầu vò từ phía x và phí o luân phiên
nếu đầu vào từ phía x là (0,0)
na trận sẽ trở thành
[[x,99,99],
 [99,99,99],
 [99,99,99]]

nếu đầu vào từ phía o là (0,0)
thông báo họ nhập lại
nếu không thì điền vào ma trận
Thử thách: nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi

```python
# Khởi tạo ma trận 3x3 toàn 99
matrix = [[99, 99, 99],
          [99, 99, 99],
          [99, 99, 99]]

current_player = 1  # X = 1, O = 0

# Hàm in ma trận
def print_matrix(m):
    for row in m:
        print(row)
    print()

# Hàm kiểm tra X thắng (3 số 1 liên tiếp)
def check_win(m):
    # Kiểm tra hàng
    for row in m:
        if row == [1, 1, 1]:
            return True

    # Kiểm tra cột
    for col in range(3):
        if m[0][col] == m[1][col] == m[2][col] == 1:
            return True

    # Kiểm tra đường chéo
    if m[0][0] == m[1][1] == m[2][2] == 1:
        return True
    if m[0][2] == m[1][1] == m[2][0] == 1:
        return True

    return False

# Vòng lặp trò chơi
while True:
    print_matrix(matrix)

    row = int(input("Nhập dòng (0-2): "))
    col = int(input("Nhập cột (0-2): "))

    # Kiểm tra ô đã nhập chưa
    if matrix[row][col] != 99:
        print("Ô đã được nhập, vui lòng nhập lại!")
        continue

    # Điền giá trị
    matrix[row][col] = current_player

    # Kiểm tra thắng
    if current_player == 1 and check_win(matrix):
        print_matrix(matrix)
        print("Người chơi X thắng!")
        break

    # Đổi lượt
    current_player = 0 if current_player == 1 else 1
```

BTVN 04
tạo 1 mảng numpy có kích thước 150x5. giả sử mảng này chauws 150 mẫu dữ liệu của sv bao gồm chiều cao, cân nặng, tuổi, lương và điểm trung bình. tách mảng dụa trên 4 cột đầu tiên thành 1 biến có tên là X và cột cuối cùng thành y

tách x thành x_train và x_test trong đó x_train chứa 70% dữ liệu và tách y thành y_train và y_test, trong đó chứa 70% dữ liệu
tạo 10 tập dữ liệu x_train không chồng chéo nhau


```python
import numpy as np
# Tạo mảng dữ liệu 150x5 (random để mô phỏng dữ liệu sinh viên)
data = np.random.rand(150, 5)
# X: 4 cột đầu (chiều cao, cân nặng, tuổi, lương)
X = data[:, :4]

# y: cột cuối (điểm trung bình)
y = data[:, 4]
# Số mẫu train (70%)
train_size = int(0.7 * len(X))

# Trộn dữ liệu trước khi chia
indices = np.random.permutation(len(X))

X_shuffled = X[indices]
y_shuffled = y[indices]

# Chia tập train và test
X_train = X_shuffled[:train_size]
X_test = X_shuffled[train_size:]

y_train = y_shuffled[:train_size]
y_test = y_shuffled[train_size:]
# Chia X_train thành 10 tập không chồng chéo
X_train_splits = np.array_split(X_train, 10)
print("Kích thước data:", data.shape)
print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

# Kiểm tra 10 tập X_train
for i, split in enumerate(X_train_splits):
    print(f"Tập X_train {i+1}: {split.shape}")
```
